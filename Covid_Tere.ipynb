```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tere = pd.read_csv('/content/Covid_Tere_Data.csv')
tere
```

```python
grafico = px.scatter_matrix(tere, dimensions=["Semanas", "Total_Confirmados", "Novas_mortes"])
grafico.show()
```

```python
tere_plot = tere.groupby('Semanas')['Novas_mortes'].sum().reset_index()

sns.lineplot(x="Semanas", y="Novas_mortes", data=tere_plot, color='blue')
plt.xlabel("Semanas Epidemologicas",fontsize=12)
plt.ylabel("Novas Mortes",fontsize=12)
plt.title("Novas mortes confirmadas")
plt.show()
```

```python
tere_plot = tere.groupby('Semanas')['Novos_casos'].sum().reset_index()

sns.lineplot(x="Semanas", y="Novos_casos", data=tere_plot, color='red')
plt.xlabel("Semanas Epidemologicas",fontsize=12)
plt.ylabel("Novos casos",fontsize=12)
plt.title("Novos casos confirmados")
plt.show()
```

```python
grafico = px.scatter(tere, x="Novos_casos", y="Novas_mortes", title='Novos casos vs Novas mortes', color='Novas_mortes')
grafico.show()
```

```python
grafico = px.scatter_matrix(tere, dimensions=["Semanas", "Total_Confirmados", "Novas_mortes"], color='Novas_mortes')
grafico.show()
```

```python
grafico = px.scatter(tere, x="Semanas", y="Novas_mortes", title='Total: Novos casos vs Novas mortes', color='Novas_mortes')
grafico.show()
```
